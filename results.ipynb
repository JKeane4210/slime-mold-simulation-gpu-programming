```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image
import numpy as np
```

```python
print(sorted(os.listdir("frames"), key=lambda x: int(x.split("_")[-1].split(".")[0])))
```

```python
fps = 30
output_file = 'slime_vid_cpu.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_file, fourcc, fps, (400, 300))

for file in sorted(os.listdir("frames"), key=lambda x: int(x.split("_")[-1].split(".")[0])):
    image = Image.open(f"frames/{file}")
    image_np = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR) / 255 * np.array([255, 255, 255])
    image_np = image_np.astype("uint8")
    image_np = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)
    out.write(image_np)
    plt.imshow(image_np)

out.release()
cv2.destroyAllWindows()
plt.show()
```
